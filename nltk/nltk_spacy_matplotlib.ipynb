```
import matplotlib.pyplot as plt
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("My name is Billy and I am coding Python")

list1 = []
for token in doc:
    list1.append(token)
    
word_lengths = [len(w) for w in list1]
plt.hist(word_lengths)
plt.show()
```

```
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt

words = word_tokenize("My name is Billy and I am coding Python")
word_lengths = [len(w) for w in words]
plt.hist(word_lengths)
plt.show()
```
