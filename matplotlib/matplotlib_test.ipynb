```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,5,21)
x
```

```python
y1 = 2**x
y2 = 5**x
```

```python
plt.plot(x,y1)
plt.xlabel('Efforts',fontsize=15)
plt.ylabel('Earnings',fontsize=15)
plt.title('Life',fontsize=15)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r')

plt.subplot(1,2,2)
plt.plot(x,y2,'b')

```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1)

axes.set_xlabel('Effort',fontsize = 15)
axes.set_ylabel('Earning',fontsize = 15)
axes.set_title('Life',fontsize = 15)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.12,0.6,0.4,0.3])

axes1.plot(x,y2,'r')
axes1.set_xlabel('Effort',fontsize = 15)
axes1.set_ylabel('Earning',fontsize = 15)
axes1.set_title('Life',fontsize = 15)

axes2.plot(x,y1,'b')
axes2.set_xlabel('Effort2',fontsize = 15)
axes2.set_ylabel('Earning2',fontsize = 15)
axes2.set_title('Life2',fontsize = 15)
```
