```
import spacy
from spacy.matcher import Matcher
import pandas as pd
import json
import time

#!spacy download en_core_web_sm
nlp = spacy.load('en_core_web_sm')
```

```
with open('Section6/data.json' , 'r') as f:
    news_articles = json.load(f)
```

```
print(news_articles[0])
```

```
print("\n".join(art['title'] for art in news_articles))
```

```
doc2 = nlp('ClearBridge MLP and Midstream Total Return Fund In (CTR) Ex-Dividend Date Scheduled for November 19, 2021')
for token in doc2:
    print(f"{token.text:15} : {token.pos_:<10} ")
```

```
ts= "\n".join(art['title'] for art in news_articles)
```

```
doc3 = nlp(ts)
for token in doc3:
    print(f"{token.text:15} : {token.pos_:<10}")
```

```
doc3 = nlp(ts)
for token in doc3:
    print(token.text , token.pos_)
```

```
docent = nlp(ts) #doc objects of all the titles
```

```
docent.ents
```

```
from spacy import displacy
```

```
displacy.render(docent ,style = 'ent' , jupyter=True , options = {'distance' : 90})
```

```
docent #doc object
```

```
for entity in docent.ents:
    print(f"{entity.label_} , {entity.text}")
```
