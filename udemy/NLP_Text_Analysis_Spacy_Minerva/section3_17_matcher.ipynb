```
import numpy as np
import pandas as pd
import spacy
from spacy.matcher import Matcher
nlp = spacy.load('en_core_web_sm')
```

```
matcher = Matcher(nlp.vocab)
```

```
pattern =  [{"LIKE_EMAIL" : True}]
matcher.add("EMAIL" , [pattern])
```

```
text = "you can send support email to itsupport@sif.com.sg or billyaung@sif.com.sg but you can also call \
us support@98779654"
doc = nlp(text)
matches = matcher(doc)
```

```
for match_id , start,end in matches:
    print(doc[start:end])
```
