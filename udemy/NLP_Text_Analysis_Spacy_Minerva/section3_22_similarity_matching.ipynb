```
import spacy
nlp = spacy.load("en_core_web_sm")
```

```
sw_spacy = nlp.Defaults.stop_words
#print(sw_spacy)
```

```
doc1 = nlp("How do I book a vacation to Paris?")
doc2 = nlp("How do I obtain an insurance?")
doc1.similarity(doc2)
```

```
doc1 = nlp("book a vacation to Paris")
doc2 = nlp("obtain an insurance")
doc1.similarity(doc2)
```

```
t1= "How do I book a vacation to Paris?"
list1 = []

for token in t1.split():
    if token.lower() not in sw_spacy:
        list1.append(token)
print("Original text                  : " , t1)
print("Text after removing stop words : " , ' '.join(list1))
```

```
t2= "How do I obtain an insurance?"
list2 = []

for token in t2.split():
    if token.lower() not in sw_spacy:
        list2.append(token)
print("Original text                  : " , t2)
print("Text after removing stop words : " , ' '.join(list2))
```

```
words = [word for word in t1.split() if word.lower() not in sw_spacy]
new_t1 = ' '.join(words)
words = [word for word in t2.split() if word.lower() not in sw_spacy]
new_t2 = ' '.join(words)
```

```
x1 = nlp(new_t1)
x2 = nlp(new_t2)
x1.similarity(x2)
```
