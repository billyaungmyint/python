```python
import spacy
```

```python
#language model
nlp = spacy.load("en_core_web_sm")
```

```python
sentences = nlp("Gucci is a famous Italian fasion brand. Ted Baker is a well known fashion brand for teenagers. Lake Chad is shrinking.  ")
```

```python
#key = nlp("fashion")
key=nlp("brand")
```

```python
for sent in sentences.sents:
           print(sent.similarity(key))
```
