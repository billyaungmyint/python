```python
import spacy
```

```python
#language model
nlp = spacy.load("en_core_web_sm")
```

doc objects of three text chunks

```python
doc1 =  nlp("Ted Baker is a popular fashion brand. Apart from sport wear, the brand specialises in smart casuals")

doc2 = nlp("Gucci is an expensive and classical fashion label. It represents the best of Italian luxury")

doc3 = nlp(" Lake Chad's boundaries have shrunk by 90%")
```

Extract entities

```python
#doc1.ents
doc1.ents
```

```python
doc2.ents
```

```python
doc3.ents
```

```python
#ents1 = [ent.text for ent in doc1.ents]
#ents2 = [ent.text for ent in doc2.ents]
#ents3 = [ent.text for ent in doc3.ents]

ents1 = [ent.text for ent in doc1.ents]
ents2 = [ent.text for ent in doc2.ents]
ents3 = [ent.text for ent in doc3.ents]
```

```python
#ents1 = nlp(" ".join(ents1))
#ents2 = nlp(" ".join(ents2))
#ents3 = nlp(" ".join(ents3))

ents1= nlp(" ".join(ents1))
ents2 = nlp(" ".join(ents2))
ents3 = nlp(" ".join(ents3))
```
