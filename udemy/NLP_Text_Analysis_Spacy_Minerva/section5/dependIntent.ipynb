```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm") #load the english model
```

```python
doc = nlp("find a doctor to give first aid")
```

```python
for token in doc:
  if token.dep_ == "dobj":
    print(token.head.text + token.text.capitalize())
```
