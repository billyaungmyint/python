```
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")
```

```
matcher = Matcher(nlp.vocab)
```

```
doc1 = nlp("I Travelling by a travelling bus")
doc2 = nlp("I traveled by bike")
```

```
pattern = [{"POS":"PRON"} , {"TEXT" : {"REGEX": "[Tt]ravel[.*]"}}]
```

```
matcher.add("travelRegex",[pattern])
```

```
for mid,start,end in matcher(doc1):
    print(start , end,doc1[start:end])
```

```
for mid,start,end in matcher(doc2):
    print(start , end,doc2[start:end])
```

```
# https://stackoverflow.com/questions/59613898/all-matches-in-a-line-spacy-matcher
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
pattern = [{'LIKE_NUM': True}, {'ORTH': '+', "OP": "?"}, {"LOWER": {"REGEX": "(?:year|month)s?"}}]
matcher.add("Skills", [pattern])

doc = nlp("1+ years of experience in XX, 2 years of experiance in YY")

matches = matcher(doc)
for _, start, end in matches:
  print(doc[start:end].text)
```
