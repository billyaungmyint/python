### .astype() method

```python
nba = pd.read_csv('../data/nba.csv').dropna(how = 'all')
```

```python
nba['Salary'].fillna(0 , inplace = True)
nba['College'].fillna('No College' , inplace= True)
```

```python
nba.head(6)
```

```python
nba.dtypes #object = string
nba.info()
```

```python
nba['Salary'] = nba['Salary'].astype("int")
```

```python
nba.info()
```

```python
nba['Number'] = nba['Number'].astype('int')
nba['Age'] = nba['Age'].astype('int')
```

```python
nba['Position'].nunique()
```

### Sort Values with .sort_values()

```python
nba.sort_values('Name' , ascending= False).head(3)
nba.sort_values('Salary' , ascending = False , na_position='first').head(3)
```

```python
nba.sort_values(['Team','Name'] , ascending= [True , False])
```

### .sort_index() method

```python
nba.sort_values(['Number','Salary','Name'] , inplace= True)
nba.tail(3)
```

```python
nba.sort_index(ascending = False , inplace= True)
nba.head(3)
```

### .rank() method

```python
nba = nba.dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0)
nba.head(3)
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')
nba.head(3)
```

```python
nba.sort_values(by = 'Salary' , ascending=False)
```
